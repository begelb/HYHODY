```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

### Load data

```python
init_data_file = 'data/vp_ic11all.dat'
next_data_file = 'data/vp_next11all.dat'

init_data = np.loadtxt(init_data_file)
next_data = np.loadtxt(next_data_file)

X = init_data[:, [0, 1]]
Y = next_data[:, [0, 1]]

init_impact_phase = init_data[:, 0]
init_impact_velocity = init_data[:, 1]

next_impact_phase = Y[:, 0]
next_impact_velocity = Y[:, 1]
```

### Plot initial conditions (first impact) and next conditions (second impact)

```python
#plt.scatter(X[:,0], X[:,1], c='purple', s = 0.1, label='Initial Data')
plt.scatter(Y[:,0], Y[:,1], c='orange', s = 0.1, label='Next Data')
plt.xlabel('Impact Phase')
plt.ylabel('Impact Velocity')
plt.legend(markerscale=10)

# calculate the min impact phase
min_impact_phase = np.min(next_impact_phase)
print('Min Impact Phase: ', min_impact_phase)

# calculate the max impact phase
max_impact_phase = np.max(next_impact_phase)

print('Max Impact Phase: ', max_impact_phase)

# calculate the min initial impact phase

min_init_impact_phase = np.min(init_impact_phase)
print('Min Initial Impact Phase: ', min_init_impact_phase)

# calculate the max initial impact phase
max_init_impact_phase = np.max(init_impact_phase)

print('Max Initial Impact Phase: ', max_init_impact_phase)
```

### Original phase surface

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Initial Phase')
ax.set_ylabel('Initial Velocity')
ax.set_zlabel('Return Phase')
ax.scatter(X[:,0], X[:,1], Y[:, 0], c='purple', s=0.1)
```

### Mod the phase variable by $6.275 \approx 2 \pi$ and replot

```python
phase_max = 6.2828711

pos_next = np.where(Y[:,0] > 0)
neg_next = np.where(Y[:,0] < 0)
plt.scatter(Y[pos_next,0], Y[pos_next,1], c='orange', s = 0.1, label='Next Data w/ Positive Phase')

# mod the phase variable by 6.275
Y[:, 0] = Y[:, 0] % (phase_max)
plt.scatter(Y[neg_next,0], Y[neg_next,1], c='blue', s = 0.1, label='Next Data w/ Negative Phase then Modded 6.275')

plt.xlabel('Impact Phase')
plt.ylabel('Impact Velocity')
plt.legend(markerscale=10)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Initial Phase')
ax.set_ylabel('Initial Velocity')
ax.set_zlabel('Return Phase')
ax.scatter(X[:,0], X[:,1], Y[:, 0], c='purple', s=0.1)
```

### Velocity surface

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], next_impact_velocity, c='purple', s=0.1)
ax.set_xlabel('Initial Phase')
ax.set_ylabel('Initial Velocity')
ax.set_zlabel('Return Velocity')
```
