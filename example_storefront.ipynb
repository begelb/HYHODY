```python
from src.example_factory import Example
import numpy as np
from src.grid import Boxes
```

```python
# Define a sample function f
def sample_f(x, y):
    return np.sin(x) * np.cos(y), np.cos(x) * np.sin(y)

lower_bounds = [-1, -1]
upper_bounds = [1, 1]

sin = Example(lower_bounds, upper_bounds, sample_f)
sin.plot()

init_data, next_data = sin.make_grid_data(100) #230

boxes = Boxes(lower_bounds, upper_bounds, 10) #40
boxes.plot(init_data, next_data, method='persistence')
```

```python
def step_func(x, y):
    return np.heaviside(x, 0), np.heaviside(y, 0)

step = Example(lower_bounds, upper_bounds, step_func)
step.plot()

init_data, next_data = step.make_grid_data(100)

boxes = Boxes(lower_bounds, upper_bounds, 11)
boxes.plot(init_data, next_data, method='persistence')
```

```python
def circ_step(x, y):
    return np.heaviside(x**2 + y**2 - 1, 0), np.heaviside(0.25 - x**2 - y**2, 0)

circ = Example(lower_bounds, upper_bounds, circ_step)
circ.plot()

init_data, next_data = circ.make_grid_data(230)

boxes = Boxes(lower_bounds, upper_bounds, 41)
boxes.plot(init_data, next_data, method='persistence')
```

```python
def smooth_bump(x, y):
    return np.exp(10*(-x**2 - y**2)), np.exp(-x**2 - y**2)

bump = Example([-1, -1], [1, 1], smooth_bump)
bump.plot()

init_data, next_data = bump.make_grid_data(230)

boxes = Boxes(lower_bounds, upper_bounds, 40)
boxes.plot(init_data, next_data, method='persistence')
```

```python
def distorted_smooth_bump(x, y):
    return np.exp(10*(-x**2 - y**2)), np.exp(-x**2 - y**2) + 0.1*np.sin(10*x)*np.sin(10*y)

dist_bump = Example([-1, -1], [1, 1], distorted_smooth_bump)
dist_bump.plot()

init_data, next_data = dist_bump.make_grid_data(230)

boxes = Boxes(lower_bounds, upper_bounds, 40)
boxes.plot(init_data, next_data, method='persistence')
```

```python
def diagonal_ridge_smooth(x, y):
    return np.exp(10*(-x**2 - y**2)), np.exp(-x**2 - y**2) + 0.1*np.sin(10*(x-y))

ridge = Example([-1, -1], [1, 1], diagonal_ridge_smooth)
ridge.plot()

init_data, next_data = ridge.make_grid_data(230)

boxes = Boxes(lower_bounds, upper_bounds, 40)
boxes.plot(init_data, next_data, method='persistence')
```

```python
# function which is discontinuous on a parabola
def parabola_discontinuous(x, y):
    return np.heaviside(y - x**2, 0), np.heaviside(x - y**2, 0)

parabola = Example([-1, -1], [1, 1], parabola_discontinuous)
parabola.plot()

init_data, next_data = parabola.make_grid_data(230)

boxes = Boxes(lower_bounds, upper_bounds, 41)
boxes.plot(init_data, next_data, method='persistence')
```

```python
def super_rapid_bump(x, y):
    return np.exp(-100*(x**2 + y**2)), np.exp(-x**2 - y**2)

super_bump = Example([-1, -1], [1, 1], super_rapid_bump)
super_bump.plot()

init_data, next_data = super_bump.make_grid_data(230)

boxes = Boxes(lower_bounds, upper_bounds, 40)
boxes.plot(init_data, next_data, method='persistence')
```
